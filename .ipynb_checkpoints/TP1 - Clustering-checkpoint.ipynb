## 2) KMeans

```python
KMeans(n_clusters=3, init='k-means++', random_state=0).fit(data)
```

```python
for k in range(2,11):
    tps1 = time.clock()
    i = KMeans(n_clusters=k, init='k-means++', random_state=0).fit(data).n_iter_    
    tps2 = time.clock()
    print("k =",k,", n_iter =",i, ", runtime = ", round((tps2 - tps1)*1000,2),"ms")
```
